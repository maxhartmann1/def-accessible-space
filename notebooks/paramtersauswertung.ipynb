```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

grouped_df = pd.read_csv("output/grouped_das.csv")

n_total_params = grouped_df["parameters"].nunique()
print(f"Anzahl verschiedener Parameterkombinationen: {n_total_params}")
```

```python
param_counts = (
    grouped_df.groupby(["match_id", "player"])["parameters"]
    .nunique()
    .reset_index(name="param_count")
)

complete_combinations = param_counts[param_counts["param_count"] == n_total_params]
display(f"Vollständige Kombinationen: {len(complete_combinations)}")
display(complete_combinations)
```

```python
fair_grouped_df = grouped_df.merge(complete_combinations, on=["match_id", "player"])

plt.figure(figsize=(12,6))
sns.boxplot(
    data=fair_grouped_df,
    x="parameters",
    y="DAS_potential_percentage",
    hue="method"
)
plt.title("Fairer Vergleich der Parameterkombinationen nach Methode")
plt.ylabel("Verbesserung (%)")
plt.xlabel("Parameterkombination")
plt.xticks(rotation=45)
plt.legend(title="Methode")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,6))
sns.barplot(
    data=fair_grouped_df,
    x="parameters",
    y="DAS_potential_percentage",
    hue="method"
)
plt.title("Vergleich der Parameterkombinationen nach Methode")
plt.ylabel("Verbesserung (%)")
plt.xlabel("Parameterkombination")
plt.xticks(rotation=45)
plt.legend(title="Methode")
plt.tight_layout()
plt.show()

```
