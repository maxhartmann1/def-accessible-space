```python
# Positionen Ermitteln -> Check mutltiple Positionen fr einen Spieler nach Mapping
position_mapping = {
    "TW": "Torwart",
    
    "IVR": "Innenverteidiger",
    "IVZ": "Innenverteidiger",
    "IVL": "Innenverteidiger",
    "RV": "Aussenverteidiger",
    "LV": "Aussenverteidiger",
    "DRM": "Aussenverteidiger",
    "DLM": "Aussenverteidiger",
    
    "DMR": "Defensives Mittelfeld",
    "DMZ": "Defensives Mittelfeld",
    "DML": "Defensives Mittelfeld",
    "RM": "Offensive Aussenbahn",
    "LM": "Offensive Aussenbahn",
    "HR": "Offensives Mittelfeld",
    "HL": "Offensives Mittelfeld",
    "ZO": "Offensives Mittelfeld",
    "ORM": "Offensive Aussenbahn",
    "OLM": "Offensive Aussenbahn",
    
    "RA": "Strmer",
    "LA": "Strmer",
    "STR": "Strmer",
    "STZ": "Strmer",
    "STL": "Strmer"
}

players["position_label"] = players["position"].map(position_mapping)
 
df_positions_per_player = (
    players.groupby(["id", "full_name", "team"])["position_label"]
    .nunique()
    .reset_index(name="num_positions")
)
players_multi = df_positions_per_player[df_positions_per_player["num_positions"] > 1]

df_position_counts = (
    players.groupby(["id", "full_name", "position_label", "team"])["game_id"]
    .nunique()
    .reset_index(name="count_games")
)
df_problem_positions = df_position_counts.merge(
    players_multi[["id", "full_name", "team"]],
    on=["id", "full_name"],
    how="inner"
)
df_main_position = (
    df_position_counts
    .sort_values(["id", "count_games"], ascending=[True, False])
    .groupby(["id", "full_name", "team"])
    .head(1)
    .reset_index(drop=True)
)
# df_main_position
```

```python
# Gruppierung pro Spieler / Spiel
df_gp = (
    players.groupby(["id", "game_id"]).agg(
        Summe_DAS_Potential=("DAS_potential", "sum"),
        Summe_DAS_Original=("DAS", "sum"),
        Mean_DAS_Potential=("DAS_potential", "mean"),
        Median_DAS_Potential=("DAS_potential", "median"),
        Median_DAS_Potential_Perc=("DAS_potential_percentage", "median"),
        n_frames=("DAS_potential", "count")
    ).reset_index()
)

df_player = (
    df_gp.groupby("id").agg(
        Total_DAS_Potential=("Summe_DAS_Potential", "sum"),
        Total_DAS_Original=("Summe_DAS_Original", "sum"),
        
        Median_DAS_Typical=("Median_DAS_Potential", "median"),
        Median_DAS_Typical_Perc=("Median_DAS_Potential_Perc", "median"),
        
        Total_Frames=("n_frames", "sum")
    ).reset_index()
)

df_player["DAS_Potential_Perc"] = df_player["Total_DAS_Potential"] / df_player["Total_DAS_Original"]
df_player["DAS_Potential"] = df_player["Total_DAS_Potential"] / df_player["Total_Frames"]


df_global_medians = (
    players.groupby("id").agg(
        Median_DAS_Total=("DAS_potential", "median"),
        Median_DAS_Total_Perc=("DAS_potential_percentage", "median")
    ).reset_index()
)
df_player = df_player.merge(df_global_medians, on="id", how="left")

df_player = df_player.merge(df_main_position, on="id", how="left")
# display(df_player)

df_compare_position = df_player.groupby("position_label")["DAS_Potential"].mean()
display(df_compare_position)
# df_gp
```

```python
def ranking_by_position_group(df, metric, min_frames=20, ascending=False):
    df = df[df["Total_Frames"] >= min_frames].copy()
    ranked = (
        df.sort_values(["position_label", metric], ascending=[True, ascending])
        .copy()
    )
    
    ranked["rank"] = ranked.groupby("position_label")[metric].rank(
        method="dense",
        ascending=ascending
    )
    return ranked

df_rank_global_percent = ranking_by_position_group(
    df_player, metric="DAS_Potential_Perc", ascending=True
)
display(df_rank_global_percent)

df_rank_global_percent.to_csv("output/rangliste.csv")
```

```python
# Gruppierung pro Spieler / Spiel
def ranking_all_by_position_group(df, metric, min_frames=20, ascending=False):
    df = df[df["n_frames"] >= min_frames].copy()
    ranked = (
        df.sort_values(["position_label", metric], ascending=[True, ascending])
        .copy()
    )
    
    ranked["rank"] = ranked.groupby("position_label")[metric].rank(
        method="dense",
        ascending=ascending
    )
    return ranked

df_group_game = df_gp.copy()

df_group_game["DAS_Potential_Perc"] = df_group_game["Summe_DAS_Potential"] / df_group_game["Summe_DAS_Original"]
df_group_game["DAS_Potential"] = df_group_game["Summe_DAS_Potential"] / df_group_game["n_frames"]

df_positions = players[["id", "game_id", "full_name", "position_label"]].drop_duplicates().copy()

df_group_game = df_group_game.merge(df_positions, on=["id", "game_id"], how="left")

df_gp_ranked = ranking_all_by_position_group(df_group_game, metric="DAS_Potential_Perc", ascending=True)
df_gp_ranked.to_csv("output/rangliste_ungrouped.csv")

df_gp_ranked
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

METRICS = [
    "Median_DAS_Typical",
    "Median_DAS_Typical_Perc",
    "DAS_Potential_Perc",
    "DAS_Potential",
    "Median_DAS_Total",
    "Median_DAS_Total_Perc",
]

def plot_top_flop_by_position(df, metric, min_frames=20, top_n=10, flop=True):
    df = df[df["Total_Frames"] >= min_frames].copy()
    groups = df["position_label"].unique()
    
    for group in groups:
        df_group = df[df["position_label"] == group].copy()
        
        if df_group.empty or group == "Torwart":
            continue
        
        df_top = df_group.nsmallest(top_n, metric)
        df_flop = df_group.nlargest(top_n, metric)
        
        plt.figure(figsize=(10, 6))
        sns.barplot(
            data=df_top,
            x=metric,
            y="full_name"
        )
        plt.title(f"Top {top_n}  {group} ({metric})")
        plt.xlabel(metric)
        plt.ylabel("Spieler")
        plt.tight_layout()
        plt.show()
        
        if flop:
            plt.figure(figsize=(10, 6))
            sns.barplot(
                data=df_flop,
                x=metric,
                y="full_name"
            )
            plt.title(f"Flop {top_n}  {group} ({metric})")
            plt.xlabel(metric)
            plt.ylabel("Spieler")
            plt.tight_layout()
            plt.show()
            
def plot_top_by_metric(df, position, min_frames=20, top_n=10):
    df = df[df["Total_Frames"] >= min_frames].copy()

    for metric in METRICS:
        df_group = df[df["position_label"] == position].copy()
        
        df_top = df_group.nsmallest(top_n, metric)
        df_flop = df_group.nlargest(top_n, metric)
        
        plt.figure(figsize=(10, 6))
        sns.barplot(
            data=df_top,
            x=metric,
            y="full_name"
        )
        plt.title(f"Top {top_n}  {position} ({metric})")
        plt.xlabel(metric)
        plt.ylabel("Spieler")
        plt.tight_layout()
        plt.show()
            

plot_top_flop_by_position(df_player, metric="DAS_Potential_Perc", top_n=1000, flop=False)    
# plot_top_flop_by_position(df_player, metric="DAS_Potential", top_n=1000, flop=False)
```
